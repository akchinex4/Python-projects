```python
## Import relevant libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline ## %matplotlib inline means to Enable inline plotting
matplotlib.rcParams['figure.figsize'] = (12,10) 



```

```python
## Read in the data
df = pd.read_csv(r'C:\Users\Willkommen\Downloads\Data_Kaggle\movies.csv')
df
```

```python
## Checking the first five rows
df.head()
```

```python
## Checking the whole datam using quick calculations
df.describe()
```

```python
## data cleaning procesing; checking for missing data

for col in df.columns:
    missing_pct = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,missing_pct))
```

```python
##data cleaning processing; correcting the year column

df['correctyear'] = df['released'].astype(str).str[:4]
df
```

```python
##data cleaning; sorting by the highest gross rATE
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
##Data cleaning; removing duplicates & showing unique values

df['company'].drop_duplicates().sort_values
```

```python
##Definning the dependent and independent variables

y = df['gross']
x = df['budget']
```

```python
##Exploring the dta using scatter plot for correlation

plt.scatter(x,y)
plt.xlabel ('Gross Earnings')
plt.ylabel ('Budget for Film')
plt.title('Gross Earnings vs Budget for Films')
plt.show()
```

```python
##Exploring the data using regression using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "green"}, line_kws={"color": "blue"})
```

```python
##Looking at the correlation

df.corr()
```

```python
##Visualizing the correlation using correlation matrix
df
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.xlabel ('Movie Features')
plt.ylabel ('Movie Features')
plt.title('Correlation Matric for Numeric Features')
plt.show()
```

```python
##Changing all categorical columns such as company, country, director, genre etc, to numerical columns using the for loop

df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
##Viewing all the columns

pd.set_option('display.max_rows', None)
```

```python
##Visualizing the correlation using correlation matrix by using the generated df_numerized

correlation_matrix = df_numerized.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.xlabel ('Movie Features')
plt.ylabel ('Movie Features')
plt.title('Correlation Matric for Numeric Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
##Unstacking the correlation for better visualization

correlation_mat = df_numerized. corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
##Sorting the correlated pairs

sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
##Determining the high correlation

high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
